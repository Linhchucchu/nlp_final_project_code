```python
import os
import pandas as pd
from google.colab import drive
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
import joblib
```

```python
drive.mount('/content/drive')

BASE_DIR = '/content/drive/MyDrive/NLP_Project_Final/Model'
DATA_DIR = os.path.join(BASE_DIR, 'SplitData')

WEIGHTS_DIR = os.path.join(BASE_DIR, 'Weights')
if not os.path.exists(WEIGHTS_DIR):
    os.makedirs(WEIGHTS_DIR)
```

```python
train_df = pd.read_csv(os.path.join(DATA_DIR, 'train_data.csv'))
test_df = pd.read_csv(os.path.join(DATA_DIR, 'test_data.csv'))

train_df = train_df.dropna()
test_df = test_df.dropna()

X_train, y_train = train_df['text'], train_df['label']
X_test, y_test = test_df['text'], test_df['label']

print(f">>> [LOADED] Train: {len(X_train)}, Test: {len(X_test)}")
```

```python
tfidf = TfidfVectorizer(max_features=10000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)

y_pred = nb_model.predict(X_test_tfidf)

label_names = ['Case Report', 'Meta-Analysis', 'Observational Study', 'Randomized Controlled Trial']

print("\n[REPORT] Baseline Performance:")

print(classification_report(y_test, y_pred, target_names=label_names))

joblib.dump(nb_model, os.path.join(WEIGHTS_DIR, 'baseline_nb_model.pkl'))
joblib.dump(tfidf, os.path.join(WEIGHTS_DIR, 'tfidf_vectorizer.pkl'))

print(f"\n>>> [SUCCESS] Baseline model and TF-IDF saved to: {WEIGHTS_DIR}")
```
